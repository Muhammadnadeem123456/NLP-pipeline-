# To tackle outlires findout min max threshold

```python
min_threshold, max_threshold= df2['price_per_sqft'].quantile([0.001, 0.999])
min_threshold, max_threshold
```

# data points less than min_threshold basically outlires

```python
df2[df2['price_per_sqft']<min_threshold]
```

```python
df2[df2['price_per_sqft']>max_threshold]
```

# removing outliers from data set

```python
df3=df2[(df2['price_per_sqft'] < max_threshold) & (df2['price_per_sqft'] > min_threshold)]
```
